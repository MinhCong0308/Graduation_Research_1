```python
import pandas as pd
df_train = pd.read_json('is_train.json')
df_valid = pd.read_json('is_val.json')
df_test = pd.read_json('is_test.json')
```

```python
x_train = df_train[0].values
y_train = df_train[1].values
x_valid = df_valid[0].values
y_valid = df_valid[1].values
x_test = df_test[0].values
y_test = df_test[1].values
```

```python
y_train = np.vectorize(label_to_num.get)(y_train)
y_valid = np.vectorize(label_to_num.get)(y_valid)
y_test = np.vectorize(label_to_num.get)(y_test)
```

```python
data_train = {'message' : x_train, 'intent' : y_train}
data_valid = {'message' : x_valid, 'intent' : y_valid}
data_test = {'message' : x_test, 'intent' : y_test}
```

```python
dataset_train = pd.DataFrame(data_train)
dataset_valid = pd.DataFrame(data_valid)
dataset_test = pd.DataFrame(data_test)
```

```python
length = [len(sentence.split()) for sentence in dataset_train['message']]
max_length = max(length)
print(max_length)
# max length = 28 in data_train
```

```python
def tokenize_function(examples):
    return tokenizer(examples["message"], padding="max_length", truncation=True, max_length=28)
```

```python
# Tokenize datasets
train_encodings = tokenizer(list(dataset_train['message']), truncation=True, padding=True)
val_encodings = tokenizer(list(dataset_valid['message']), truncation=True, padding=True)
test_encodings = tokenizer(list(dataset_test['message']), truncation=True, padding=True)
```

```python
print(train_encodings[0])
```

```python
import torch
# convert datasets to pytorch tensors
class IntentDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item
    def __len__(self):
        return len(self.labels)
```

```python
train_dataset = IntentDataset(train_encodings, list(dataset_train['intent']))
val_dataset = IntentDataset(val_encodings, list(dataset_valid['intent']))
test_dataset = IntentDataset(test_encodings, list(dataset_test['intent']))
```

```python
from transformers import DataCollatorWithPadding
```

```python
class MT5EncoderForIntentRecognition(torch.nn.Module):
    def __init__(self, encoder, num_labels):
        super(MT5EncoderForIntentRecognition, self).__init__()
        self.encoder = encoder
        self.dropout = torch.nn.Dropout(0.1)
        self.classifier = torch.nn.Linear(self.encoder.config.d_model, num_labels)

    def forward(self, input_ids, attention_mask=None, labels=None):
        encoder_outputs = self.encoder(input_ids=input_ids, attention_mask=attention_mask)
        sequence_output = encoder_outputs.last_hidden_state
        pooled_output = sequence_output[:, 0, :]
        pooled_output = self.dropout(pooled_output)
        logits = self.classifier(pooled_output)

        loss = None
        if labels is not None:
            loss_fct = torch.nn.CrossEntropyLoss()
            loss = loss_fct(logits.view(-1, self.classifier.out_features), labels.view(-1))

        return {"loss": loss, "logits": logits} if loss is not None else {"logits": logits}
```

```python
num_labels = len(unique_labels)
model = MT5EncoderForIntentRecognition(encoder,num_labels)
```

```python
from transformers import Trainer, TrainingArguments
```

```python
training_args = TrainingArguments(
    output_dir='./results',
    evaluation_strategy='epoch',
    save_strategy='epoch',
    learning_rate=2e-4,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    num_train_epochs=5,
    weight_decay=0.01,
    logging_dir='./logs',
    load_best_model_at_end = True
)

trainer = Trainer(
    model = model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=tokenizer,
    data_collator=DataCollatorWithPadding(tokenizer=tokenizer)
)
```

```python
for name, param in model.named_parameters():
    if not param.is_contiguous():
        param.data = param.contiguous()
trainer.train()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)

    accuracy = accuracy_score(labels, preds)
    precision = precision_score(labels, preds, average='weighted', zero_division=0)
    recall = recall_score(labels, preds, average='weighted', zero_division=0)
    f1 = f1_score(labels, preds, average='weighted', zero_division=0)

    return {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1
    }
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    eval_dataset=test_dataset,
    tokenizer=tokenizer,
    data_collator=DataCollatorWithPadding(tokenizer=tokenizer),
    compute_metrics=compute_metrics
)

# Run evaluation
results = trainer.evaluate()
print(results)
```

```python
import torch
from transformers import AutoConfig, AutoTokenizer, MT5Model

class MT5EncoderForIntentRecognition(torch.nn.Module):
    def __init__(self, encoder, num_labels):
        super(MT5EncoderForIntentRecognition, self).__init__()
        self.encoder = encoder
        self.dropout = torch.nn.Dropout(0.1)
        self.classifier = torch.nn.Linear(self.encoder.config.d_model, num_labels)

    @classmethod
    def from_pretrained(cls, model_name, num_labels):
        config = AutoConfig.from_pretrained(model_name)
        encoder = MT5Model.from_pretrained(model_name).encoder

        model = cls(encoder, num_labels)
        state_dict = torch.load(os.path.join(model_name, "pytorch_model.bin"), map_location=torch.device('cuda' if torch.cuda.is_available() else 'cpu'))
        model.encoder.load_state_dict(state_dict['encoder_state_dict'], strict=False)
        model.classifier.load_state_dict(state_dict['classifier_state_dict'], strict=False)

        return model

    def forward(self, input_ids, attention_mask=None, labels=None):
        encoder_outputs = self.encoder(input_ids=input_ids, attention_mask=attention_mask)
        sequence_output = encoder_outputs.last_hidden_state
        pooled_output = sequence_output[:, 0, :]
        pooled_output = self.dropout(pooled_output)
        logits = self.classifier(pooled_output)

        loss = None
        if labels is not None:
            loss_fct = torch.nn.CrossEntropyLoss()
            loss = loss_fct(logits.view(-1, self.classifier.out_features), labels.view(-1))

        return {"loss": loss, "logits": logits} if loss is not None else {"logits": logits}
```

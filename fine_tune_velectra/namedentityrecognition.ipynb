```python
import pandas as pd
from datasets import Dataset


file_path = '/content/ner_dataset.conll'

data = []
with open(file_path, 'r', encoding='utf-8') as file:
    sentence = []
    ner_tags = []
    i = 0
    for line in file:
        line = line.strip()
        i += 1
        if line:
          parts = line.split()
          if len(parts) == 2:
            word, tag = parts
            sentence.append(word)
            ner_tags.append(tag)
          else:
            print(f"Skipping line due to unexpected format: {i}") #print line be error in dataset
        else:
            data.append({'words': sentence, 'ner_tags': ner_tags})
            sentence = []
            ner_tags = []

dataset = Dataset.from_list(data)

```

```python
def encode_labels(ner_tags):
    return [label_to_id[tag] for tag in ner_tags]
def tokenize_and_align_labels(examples):
    # Tokenize the input words (batched)
    tokenized_inputs = tokenizer(examples['words'],
                                 is_split_into_words=True,  # Since we're passing a list of word lists
                                 padding='max_length',
                                 truncation=True,
                                 max_length=128,
                                 add_special_tokens=True)

    labels = []
    for i, label in enumerate(examples['ner_tags']):
        # Map sequence to numeric labels
        encoded_labels = encode_labels(label)

        word_ids = tokenized_inputs.word_ids(batch_index=i)
        label_ids = []
        previous_word_idx = None

        for word_idx in word_ids:
            if word_idx is None:
                label_ids.append(-100)  # Padding token, ignore
            elif word_idx != previous_word_idx:
                label_ids.append(encoded_labels[word_idx])  # Map only the first subword
            else:
                label_ids.append(-100)  # Ignore subwords

            previous_word_idx = word_idx

        labels.append(label_ids)

    # Return labels in tokenized_inputs
    tokenized_inputs['labels'] = labels

    return tokenized_inputs

```

```python
tokenized_dataset = dataset.map(tokenize_and_align_labels, batched=True)
```

```python
from datasets import DatasetDict

# Step 1: Split the dataset into train (70%) and test (30%)
split_dataset = tokenized_dataset.train_test_split(test_size=0.2)

# Step 2: Take 15% of the training set to be the validation set
train_val_dataset = split_dataset['train'].train_test_split(test_size=0.15)

# Step 3: Create a DatasetDict to store the train, validation, and test sets
dataset_dict = DatasetDict({
    'train': train_val_dataset['train'],        # 85% of the original train set (70% of the full dataset)
    'validation': train_val_dataset['test'],    # 15% of the original train set (10.5% of the full dataset)
    'test': split_dataset['test']               # 30% of the full dataset
})

print(dataset_dict)
```

```python
from collections import Counter

# Initialize a Counter to hold label counts
label_counts = Counter()

# Iterate through all splits in the dataset (train, validation, and test)
for split in ['train', 'validation', 'test']:
    for example in dataset_dict[split]:  # Iterate through each example in the split
        ner_tags = example['ner_tags']
        label_counts.update(ner_tags)

# Display the counts for each label
for label, count in label_counts.items():
    print(f"{label}: {count}")
```

```python
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

def compute_metrics(p):
    preds = p.predictions.argmax(-1)
    labels = p.label_ids
    mask = labels != -100
    precision, recall, f1, _ = precision_recall_fscore_support(labels[mask], preds[mask], average='weighted')
    acc = accuracy_score(labels[mask], preds[mask])
    return {
        'accuracy': acc,
        'f1': f1,
        'precision': precision,
        'recall': recall,
    }
```

```python
from transformers import AutoModelForTokenClassification, TrainingArguments, Trainer

# Load the model for token classification (NER task)
model = AutoModelForTokenClassification.from_pretrained("FPTAI/velectra-base-discriminator-cased", num_labels=len(label_to_id))

# Set training arguments
training_args = TrainingArguments(
    output_dir='./results',            # Directory for saving model checkpoints
    eval_strategy="epoch",             # Evaluate at the end of each epoch
    save_strategy="epoch",             # Match the save strategy to evaluation
    logging_dir='./logs',              # Directory for logging
    logging_steps=10,                  # Log every 10 steps
    per_device_train_batch_size=4,     # Batch size per device for training
    per_device_eval_batch_size=4,      # Batch size per device for evaluation
    num_train_epochs=7,               # Number of training epochs
    weight_decay=0.01,                 # Weight decay for optimization
    learning_rate=7e-5,                # Adjusted learning rate
    load_best_model_at_end=True,       # Load the best model at the end of training
)

# Initialize Trainer object
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset_dict['train'],
    eval_dataset=dataset_dict['validation'],
    tokenizer=tokenizer,
    compute_metrics = compute_metrics,
)

# Train the model
trainer.train()
```

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
import torch
from nltk.tokenize import word_tokenize


tokenizer = AutoTokenizer.from_pretrained("NguyenMinh03082004/velectra_fine_tune_for_laptop_ner")
model = AutoModelForTokenClassification.from_pretrained("NguyenMinh03082004/velectra_fine_tune_for_laptop_ner")

# Set model to evaluation mode
model.eval()

# Example sentence for NER
text = "Tôi cần mua laptop tầm giá 25 triệu để làm về mô hình AI"
words = word_tokenize(text)  # Tokenize sentence into words

# Tokenize the text with word-level tokenization enabled
inputs = tokenizer(words, is_split_into_words=True, return_tensors='pt', padding=True, truncation=True, max_length=128)

# Perform inference
with torch.no_grad():
    outputs = model(**inputs)

id_to_label = {idx: label for label, idx in label_to_id.items()}
# Convert predictions to label names
predictions = torch.argmax(outputs.logits, dim=-1)



# Convert tokens and labels
tokens = tokenizer.convert_ids_to_tokens(inputs['input_ids'][0])
predicted_labels = [id_to_label[label_id] for label_id in predictions[0].cpu().numpy()]

# Combine subword tokens and their predicted labels
combined_tokens = []
combined_labels = []
current_token = ""
current_label = ""

for token, label in zip(tokens, predicted_labels):
    if token.startswith("##"):
        current_token += token[2:]  # Append subword (remove '##')
    elif token in ["[CLS]", "[SEP]", "[PAD]"]:  # Ignore special tokens
        continue
    elif current_token and (token.startswith(".")  or current_token.isdigit() and token.isdigit()): # check for current token is digit and next token still is digit
            current_token += token
    else:
        if current_token:  # Add the last token and label
            combined_tokens.append(current_token)
            combined_labels.append(current_label)
        current_token = token
        current_label = label

# Finalize the last token
if current_token:
    combined_tokens.append(current_token)
    combined_labels.append(current_label)

# Print tokens with predicted labels
for token, label in zip(combined_tokens, combined_labels):
    print(f"{token}: {label}")
```

```python
# Set model to evaluation mode
model.eval()

# Initialize a list to hold the results for all test examples
all_results = []

for example in dataset_dict['test']:
    text = " ".join(example['words'])  # Combine words to form the full sentence
    inputs = tokenizer(text, padding=True, truncation=True, max_length=128, return_tensors='pt')

    with torch.no_grad():
        outputs = model(**inputs)

    predictions = torch.argmax(outputs.logits, dim=-1)
    predicted_labels = [id_to_label[label_id] for label_id in predictions[0].numpy()]
    tokens = tokenizer.convert_ids_to_tokens(inputs['input_ids'][0])

    combined_tokens = []
    combined_labels = []
    current_token = ""
    current_label = ""

    for token, label in zip(tokens, predicted_labels):
        if token.startswith("##"):
            current_token += token[2:]  # Skip the '##' and append the subword
        elif current_token and (token.startswith(".") or current_token.isdigit() and token.isdigit()):
            current_token += token
        else:
            if current_token:
                combined_tokens.append(current_token)
                combined_labels.append(current_label)
            current_token = token
            current_label = label

    if current_token:
        combined_tokens.append(current_token)
        combined_labels.append(current_label)

    all_results.append({
        'tokens': combined_tokens,
        'labels': combined_labels
    })

for result in all_results:
    print("Tokens and predicted labels:")
    for token, label in zip(result['tokens'], result['labels']):
        print(f"{token}: {label}")
    print("\n")  # Add a newline for separation between examples
```
